```python
import numpy as np #kalkulasi least square
import matplotlib.pyplot as plt # utk menggambar suatu grafik hasil perhitungan
```

```python
# masukkan data set
#least square menggunakan methode direct inverse
x = np.array([90, 100, 120, 150, 200])
y = np.array([2, 3, 10, 15, 20])

print(x)
print(y)
```

```python
#mencari nilai A
A = np.vstack([x, np.ones(len(x))]).T

print(A)
```

```python
# simbol : -> slicing -> utk mengambil array dr index tertentu sampai ke-n-1
y = y[:, np.newaxis]
# slicing

# print(y[0:4])

print(y)
```

```python
# (A^T*A)^-1*A^T*Y
# direct inverse -> menghitung rumus secara langsung
# alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T, A)), A.T)),y)

# pseudo inverse
# (A^T*A)^-1*A^T -> pseudo inverse
pinv = np.linalg.pinv(A)
# pseudo inverse *y
alpha = pinv.dot(y)
print(alpha)
```

```python
# bikin grafik
plt.figure(figsize=(10, 8))
plt.plot(x, y, 'co')

# persamaan garisnya
# y = mx + b
plt.plot(x, alpha[0]*x + alpha[1], 'm')

plt.show()
```
