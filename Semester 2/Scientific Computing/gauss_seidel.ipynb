```python
import numpy as np

def isDiagonallyDominant(x):

    x = np.array(x)

    diag = np.diag(np.abs(x))

    off_diag = np.sum(np.abs(x), axis = 1) - diag

    if np.all(diag > off_diag):
        return True
    else:
        return False 
    
def gauss_seidel(x, y, epsilon = 0.01, t = 15):
    if not isDiagonallyDominant(x):
        print('is not diagonally dominant')
        return
    
    x = np.array(x)
    y = np.array(y)

    #mengambil nilai diagonal
    diag = np.diag(np.abs(x))

    #negasi nilai x
    x = -x

    #ngisi diagonalnya dengan 0
    np.fill_diagonal(x, 0)

    mtx_old = np.zeros(x[0].shape)

    #dia bakal ngukur sesuai dengan batas iterasi
    for i in range(t):
        mtx_new = np.array(mtx_old)
        #mengisi/menghitung setiap data iterasi
        for idx, row in enumerate(x):
            mtx_new[idx] = (y[idx] + np.dot(row, mtx_new)) / diag[idx]

        print(f'iterasi ke - {i}')
        print(mtx_new)

        dx = np.sqrt(np.dot(mtx_new - mtx_old, mtx_new - mtx_old))
        print(f'dx : -{dx}')

        if dx < epsilon:
            print('converged')
            return
        
        mtx_old = mtx_new

    print('not converged')
    return

Xs = [[10, 4, 5],
      [1, 6, 1],
      [3, 1, 6]]

Ys = [8, 5, 10]

gauss_seidel(Xs, Ys)
```
